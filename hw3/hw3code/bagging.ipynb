```python
import scipy.io as sio
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import scale
import lightgbm as lgb
```

```python
mat_contents = sio.loadmat('MSdata.mat')
```

```python
X_train = mat_contents['trainx']
y_train = mat_contents['trainy']
X_test = mat_contents['testx']
```

```python
##split the data to train and validation set.
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
##set the parameters of the model
model_lgb = lgb.LGBMRegressor(objective='regression',num_leaves=800,
                              learning_rate=0.05, n_estimators=15000,
                              max_bin = 55, bagging_fraction = 0.8,
                              bagging_freq = 10, feature_fraction = 0.2319,
                              min_data_in_leaf =200, min_sum_hessian_in_leaf = 11)
```

```python
##train the model
model_lgb.fit(X_train, y_train.ravel(), eval_set = [(X_val, y_val.ravel())], early_stopping_rounds = 50, eval_metric = 'mae' )
```

```python
##predict the value
y_pred = model_lgb.predict(X_test)
```

```python
##calculate the MAE score
mean_absolute_error(y_pred, y_val)
```

```python
## output pred value to csv
dataid = np.array(list(range(1, 51631)))
dataid.tolist()
res = {'dataid': dataid, 'prediction': y_pred}
res = pd.DataFrame(res)
res.to_csv('lgb.csv', index = False)
```
